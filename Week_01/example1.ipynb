### functions, loops, control flow 

````python
def myFunction(a, b):
    for num1 in range(a):
        if num1 % 2 == 0 and num1 % 4 == 0:
            print(str(num1) + " is multiple of four!")
        elif num1 % 2 == 0 and num1 % 4 != 0:
            print(str(num1) + " is even, but not a multiple of four!")
        else:
            print(str(num1) + " is odd!")
    for num2 in range(1, b, 2):
        print(num2, 2**num2)
````

````python
a = 5
b = 10
myFunction(a, b)
````

### üîπ –ó–∞–¥–∞—á–∞ 3: –†–µ—á–Ω–∏–∫ –æ—Ç –¥–≤–∞ —Å–ø–∏—Å—ä–∫–∞

–î–∞–¥–µ–Ω–∏ —Å–∞ –¥–≤–∞ —Å–ø–∏—Å—ä–∫–∞:

```python
keys = [1, 2, 3, 4, 5]
words = ["one", "two", "three", "four", "five"]
```

–°—ä–∑–¥–∞–π—Ç–µ —Ä–µ—á–Ω–∏–∫, –≤ –∫–æ–π—Ç–æ:

- –≤—Å–µ–∫–∏ –∫–ª—é—á –∏–¥–≤–∞ –æ—Ç `keys`
- –≤—Å—è–∫–∞ —Å—Ç–æ–π–Ω–æ—Å—Ç –µ –Ω–∞—Ä–µ–¥–µ–Ω–∞ –¥–≤–æ–π–∫–∞:
  - –∫–ª—é—á–∞ –Ω–∞ —Ç—Ä–µ—Ç–∞ —Å—Ç–µ–ø–µ–Ω
  - —Å—ä–æ—Ç–≤–µ—Ç–Ω–∞—Ç–∞ –¥—É–º–∞ –æ—Ç `words`

–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏ –º–æ–∂–µ–º –¥–∞ –≥–æ –∑–∞–ø–∏—à–µ–º –∫–∞—Ç–æ:

$$
D = \{\, keys_i \mapsto (keys_i^3, words_i) \;\mid\; i \in \{1,2,3,4,5\} \}
$$



````python
# solution
keys = [1, 2, 3, 4, 5]
words = ["one", "two", "three", "four", "five"]

D = {keys[i]: (keys[i]**3, words[i]) for i in range(len(keys))}
print(D)
````

````python
# solution - alternative
D_alt = {k: (k**3, w) for k, w in zip(keys, words)}
print(D_alt)
````

````python
for i, k in enumerate(words):
    print(i, " :: ", k)
````
